```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Q3

```python
fig3  = plt.figure()
plt.plot(range(1,13),df['total_profit'])
plt.xlabel('Month')
plt.ylabel('Profit')
plt.xticks(range(1,13))
fig.show()
```

#### Q4

```python
fig3.subplots(2)
plt.plot(range(1,13),df['total_profit'],linestyle='dotted', color='red', marker = 'o', markerfacecolor='red', linewidth=3)
plt.legend(['Amount in Rs.'], loc = 'lower right')
plt.xlabel('Month')
plt.xticks(range(1,13))
plt.ylabel('Number of Units sold')
```
